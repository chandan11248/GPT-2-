```python
import tiktoken
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text="my name is chandan shah, i ama student."
ids=tokenizer.encode(text)
```

```python
tokenizer.decode(ids)
```

```python

text1="hi my self chandan shah"
text3=" Even if she had no idea where"
text2="i am a student."
text="<|endoftext|>".join((text1,text3,text2))
ids=tokenizer.encode(text,allowed_special={"<|endoftext|>"})
```

```python
tokenizer.decode(ids)
```
